```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a directed graph
G = nx.DiGraph()

# Add nodes for the input layer
G.add_node('$x_2$', pos=(0, 0))
G.add_node('$x_1$', pos=(0, 1))
G.add_node('$b=1$', pos=(0, 2))

# Add node for the output layer
G.add_node('$h(z)$', pos=(0.25, 1))

# Add an out arrow from the output node
G.add_node('$y$', pos=(0.5, 1))

# Add edges between the input layer nodes and the output node
G.add_edge('$x_2$', '$h(z)$', weight='$w_2$')
G.add_edge('$x_1$', '$h(z)$', weight='$w_1$')
G.add_edge('$b=1$', '$h(z)$', weight='$w_0$')

G.add_edge('$h(z)$', '$y$', style='dashed')  # Edge for the arrow

# Define positions for all nodes
pos = nx.get_node_attributes(G, 'pos')

# Draw the neural network
nx.draw(G, pos, with_labels=True, node_size=2000, node_color='lightgreen', 
        font_size=12, font_weight='bold', arrows=True, arrowstyle='-|>')

# Draw the edge labels
edge_labels = {('$x_1$', '$h(z)$'): '$w_1$', ('$b=1$', '$h(z)$'): '$w_0$', ('$x_2$', '$h(z)$'): '$w_2$'}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='red', font_size=10)

# Save the plot as an image file with higher resolution
plt.title('Single-layer Neural Network Visualization')
plt.savefig('imgs/slp.png', format='png', dpi=700)

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import networkx as nx

# Create a directed graph
G = nx.DiGraph()

# Add nodes for the input layer
G.add_node('$x_2$', pos=(0, 0))
G.add_node('$x_1$', pos=(0, 1))
G.add_node('$b=1$', pos=(0, 2))

# Add nodes for the hidden layer with 2 units
G.add_node('$6$', pos=(0.25, 0.5))
G.add_node('$0$', pos=(0.25, 1.5))

# Add node for the output layer
G.add_node('$h(z)$', pos=(0.5, 1))

# Add an out arrow from the output node
G.add_node('$y$', pos=(0.75, 1))

# Add edges between the input layer nodes and the hidden layer nodes
G.add_edge('$x_2$', '$6$', weight='$w_{2,1}$')
G.add_edge('$x_1$', '$6$', weight='$w_{1,1}$')
G.add_edge('$b=1$', '$6$', weight='$w_{0,1}$')

G.add_edge('$x_2$', '$0$', weight='$w_{2,2}$')
G.add_edge('$x_1$', '$0$', weight='$w_{1,2}$')
G.add_edge('$b=1$', '$0$', weight='$w_{0,2}$')

# Add edges between the hidden layer nodes and the output node
G.add_edge('$6$', '$h(z)$', weight='$w_{h1}$')
G.add_edge('$0$', '$h(z)$', weight='$w_{h2}$')

# Add an edge for the output node to the final output
G.add_edge('$h(z)$', '$y$', style='dashed')

# Define positions for all nodes
pos = nx.get_node_attributes(G, 'pos')

# Draw the neural network
nx.draw(G, pos, with_labels=True, node_size=2000, node_color='lightgreen',
        font_size=12, font_weight='bold', arrows=True, arrowstyle='-|>')

# Draw the edge labels

edge_labels = {
  #  ('$x_2$', '$h_1$'): '$w_{2,1}$', ('$x_1$', '$h_1$'): '$w_{1,1}$', ('$b=1$', '$h_1$'): '$w_{0,1}$',
  #  ('$x_2$', '$h_2$'): '$w_{2,2}$', ('$x_1$', '$h_2$'): '$w_{1,2}$', ('$b=1$', '$h_2$'): '$w_{0,2}$',
    ('$6$', '$h(z)$'): '$w_{h1}=1$', ('$0$', '$h(z)$'): '$w_{h2}=-1$'
}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='red', font_size=10)

# Save the plot as an image file with higher resolution
plt.title('Neural Network with Hidden Layers')
plt.savefig('imgs/chap2_ex1.png', format='png', dpi=700)

# Show plot
plt.show()
```
