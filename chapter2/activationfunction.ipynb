```python
import numpy as np
import matplotlib.pyplot as plt

# Define the step function
def step_function(x):
    return np.where(x >= 0, 1, 0)

# Generate x values
x = np.linspace(-5, 5, 100)

# Compute y values
y = step_function(x)

# Plot
plt.plot(x, y)
plt.title('Step Function')
plt.xlabel('z')
plt.ylabel('h(z)')
plt.grid(True)
plt.savefig("imgs/step_fun.png", dpi = 700)
plt.show()
```

```python
# Sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Generate x values
x = np.linspace(-5, 5, 100)

# Compute y values for sigmoid function
y_sigmoid = sigmoid(x)

# Plot sigmoid function
plt.plot(x, y_sigmoid, label='Sigmoid', color='blue')
plt.title('Sigmoid Function')
plt.xlabel('z')
plt.ylabel('h(z)')
plt.grid(True)
plt.legend()


# Save plot to an image file
plt.savefig("imgs/sigmoid_fun.png", dpi = 700)
plt.show()
```

```python
# Tanh function
def tanh(x):
    return np.tanh(x)

# Generate x values
x = np.linspace(-5, 5, 100)

# Compute y values for tanh function
y_tanh = tanh(x)

# Plot tanh function
plt.plot(x, y_tanh, label='Tanh', color='green')
plt.title('Tanh Function')
plt.xlabel('z')
plt.ylabel('h(z)')
plt.grid(True)
plt.legend()

# Save plot to an image file
plt.savefig("imgs/tanh_fun.png", dpi = 700)
plt.show()
```

```python

# ReLU function
def relu(x):
    return np.maximum(0, x)

# Generate x values
x = np.linspace(-5, 5, 100)

# Compute y values for ReLU function
y_relu = relu(x)

# Plot ReLU function
plt.plot(x, y_relu, label='ReLU', color='red')
plt.title('ReLU Function')
plt.xlabel('z')
plt.ylabel('h(z)')
plt.grid(True)
plt.legend()


# Save plot to an image file
plt.savefig("imgs/relu_fun.png", dpi = 700)

plt.show()
```

```python
# Softmax function
def softmax(x):
    exp_values = np.exp(x - np.max(x))
    return exp_values / np.sum(exp_values)

# Generate example values
x = np.linspace(-5, 5, 100)
y_softmax = softmax(x)

# Plot softmax function
plt.plot(x, y_softmax, label='Softmax', color='blue')
plt.title('Softmax Function')
plt.xlabel('z')
plt.ylabel('h(z)')
plt.grid(True)
plt.legend()

# Save plot to an image file
plt.savefig("imgs/softmax_fun.png", dpi = 700)
plt.show()

```
