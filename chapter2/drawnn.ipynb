```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a directed graph
G = nx.DiGraph()

# Add nodes for the input layer
G.add_node('$x_2$', pos=(0, 0))
G.add_node('$x_1$', pos=(0, 1))
G.add_node('$b=1$', pos=(0, 2))

# Add node for the output layer
G.add_node('$h(z)$', pos=(0.25, 1))

# Add an out arrow from the output node
G.add_node('$y$', pos=(0.5, 1))

# Add edges between the input layer nodes and the output node
G.add_edge('$x_2$', '$h(z)$', weight='$w_2$')
G.add_edge('$x_1$', '$h(z)$', weight='$w_1$')
G.add_edge('$b=1$', '$h(z)$', weight='$w_0$')

G.add_edge('$h(z)$', '$y$', style='dashed')  # Edge for the arrow

# Define positions for all nodes
pos = nx.get_node_attributes(G, 'pos')
plt.figure(figsize=(10, 6))
# Draw the neural network
nx.draw(G, pos, with_labels=True, node_size=2000, node_color='yellow',  
        font_size=12, font_weight='bold', arrows=True, arrowstyle='-|>')

# Draw the edge labels
edge_labels = {('$x_1$', '$h(z)$'): '$w_1$', ('$b=1$', '$h(z)$'): '$w_0$', ('$x_2$', '$h(z)$'): '$w_2$'}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='red', font_size=10)


# Save the plot as an image file with higher resolution
plt.title('Single-layer Neural Network Visualization')
plt.savefig('imgs/slp_eg.png', format='png', dpi=700)

# Show plot
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a directed graph
G = nx.DiGraph()

# Define the layers and nodes
input_layer = ['bias', '$x_1$', '$x_2$']
hidden_layer = ['$h(z^1_1)$', '$h(z^1_2)$']
output_layer = ['$h(z^o)$']

# Add nodes to the graph
G.add_nodes_from(input_layer)
G.add_nodes_from(hidden_layer)
G.add_nodes_from(output_layer)

# Connect input layer to hidden layer
for input_node in input_layer:
    for hidden_node in hidden_layer:
        G.add_edge(input_node, hidden_node)

# Connect hidden layer to output layer
for hidden_node in hidden_layer:
    G.add_edge(hidden_node, output_layer[0])

# Define the position of nodes for better visualization
pos = {
    'bias': (0, 2),
    '$x_1$': (0, 1),
    '$x_2$': (0, 0),
    '$h(z^1_1)$': (1, 1.5),
    '$h(z^1_2)$': (1, 0.5),
    '$h(z^o)$': (2, 1)
}

# Draw the network
plt.figure(figsize=(10, 7))
nx.draw(G, pos, with_labels=True, node_size=3000, node_color='lightgreen', arrows=True, font_size=16)


plt.title('------------------------ Forward Propagation----------------------->', size = 20)
plt.savefig('imgs/fw_eg.png', format='png', dpi=700, bbox_inches='tight')
plt.show()
```
